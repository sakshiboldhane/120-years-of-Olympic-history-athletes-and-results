```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline
```

```python
# Load dataset
athletes =  pd.read_csv('D:/Data set olympics/athlete_events.csv')
region = pd.read_csv('D:/Data set olympics/noc_regions.csv')
```

```python
athletes.head()
```

```python
region.head()
```

```python
athletes_df =athletes.merge(region,how='left',on='NOC')
athletes_df.head()
# Group the athletes by country and sport and count the number of athletes
```

```python
athletes_df.shape
```

```python
#Column names consistent
athletes_df.rename(columns={'region':'Region','notes':'Notes'},inplace=True)
```

```python
#Top countries participating 
top_10_countries=athletes_df['Team'].value_counts().sort_values(ascending=False).head(10)
top_10_countries
```

```python
#Plot for the top 10 countries
plt.figure(figsize=(12,6))
#plt.xticks(rotation=20)
plt.title('Overall participation by country')
sns.barplot(x=top_10_countries, y=top_10_countries, palette='Set2')
```

```python
#Age distribution of the participants
plt.figure(figsize=(12,6))
plt.title('Age distribution of participants')
plt.xlabel('Age')
plt.ylabel('Number of participants')
plt.hist(athletes_df.Age , bins=np.arange(10,80,2),color='orange',edgecolor='White')
```

```python
winter_sports=athletes_df[athletes_df.Season=="Winter"].Sport.unique()
winter_sports
```

```python
#Summer olympics Sports
summer_sports=athletes_df[athletes_df.Season=="Summer"].Sport.unique()
summer_sports
```

```python
#Male and female participation
male_female_ratio=athletes_df.Sex.value_counts()
male_female_ratio
```

```python
#Pie plot for male and femlale athlete
plt.figure(figsize=(10,6))
plt.title('Gender Distribution')
plt.pie(male_female_ratio,labels=male_female_ratio.index,autopct='%1.1f%%',startangle=180,shadow=True)
```

```python
#Total Medals
athletes_df.Medal.value_counts()
```

```python
#Total number of female athletes in each olymphics
female_participants=athletes_df[(athletes_df.Sex=='F')& (athletes_df.Season=='Summer')][['Sex','Year']]
female_participants=female_participants.groupby('Year').count().reset_index()
female_participants.head()
```

```python
women0lymopics=athletes_df[(athletes_df.Sex=='F')& (athletes_df.Season=='Summer')]
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=(20, 10))
sns.countplot(x='Year',data=women0lymopics, palette="Spectral")
plt.title('Women Participation')
```

```python
#Number of athletes in each sport
part=women0lymopics.groupby('Year')['Sex'].value_counts()
plt.figure(figsize=(20, 10))
part.loc[:,'F'].plot()
plt.title('Plot of female participation')
#Number of athletes in each sport
```

```python
#Gold Medal Athlete
goldMedals=athletes_df[(athletes_df.Medal=='Gold')]
goldMedals.head()

```

```python
#take only the values that are different from NaN
goldMedals=goldMedals[np.isfinite(goldMedals['Age'])]
                      
```

```python
#Gold Beyond 60
goldMedals['ID'][goldMedals['Age']>60].count()
```

```python
sporting_event=goldMedals['Sport'][goldMedals['Age']>60]
sporting_event
```

```python
#Plot for sporting events 
plt.figure(figsize=(10,5))
plt.tight_layout()
sns.countplot(sporting_event)
plt.title("Gold Medal for Athlete over 60 yeards")
```

```python
#Gold medal for each country
goldMedals.Region.value_counts().reset_index(name="Medal").head(5)
```

```python
goldmedals=goldMedals.Region.value_counts().reset_index(name='Medal').head(6)
g=sns.catplot(x="Region",y="Medal",data=goldmedals,height=5,kind="bar",palette="rocket")
g.despine(left=True)
g.set_xlabels("Top 5 countries")
g.set_ylabels("Number of Gold Medals")
plt.title("Gold medal per country")
```
